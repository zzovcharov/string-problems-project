# Check If Two String Arrays are Equivalent

## Problem
Given two string arrays `word1` and `word2`, return **true** if the two arrays represent the same string, and **false** otherwise.

An array represents a string if concatenating its elements **in order** forms that string.

### Examples
- **Input:** `word1 = ["ab", "c"]`, `word2 = ["a", "bc"]`
  **Output:** `true`
  **Reason:** `"ab" + "c" = "abc"` and `"a" + "bc" = "abc"`

- **Input:** `word1 = ["a", "cb"]`, `word2 = ["ab", "c"]`
  **Output:** `false`

- **Input:** `word1 = ["abc", "d", "defg"]`, `word2 = ["abcddefg"]`
  **Output:** `true`

### Constraints
- `1 <= word1.length, word2.length <= 10^3`
- `1 <= word1[i].length, word2[i].length <= 10^3`
- `1 <= sum(len(word1[i])), sum(len(word2[i])) <= 10^3`
- `word1[i]` and `word2[i]` contain lowercase English letters only


## Approach

1. **Concatenate** each array of strings into a single string using `"".join(...)`.
2. **Compare** the two resulting strings for equality.
3. Return the boolean result.

This is optimal here because the total number of characters is bounded (≤ 1000), so joining is efficient and simple.


```python
class Solution(object):
    def arrayStringsAreEqual(self, word1, word2):
        """
        :type word1: List[str]
        :type word2: List[str]
        :rtype: bool
        """
        word1, word2 = ''.join(word1), ''.join(word2)

        return word1 == word2
```

```python
# Enter arrays as Python list literals, e.g. ["ab","c"]
word1 = eval(input('Enter word1 as a list of strings (e.g., ["ab","c"]): '))
word2 = eval(input('Enter word2 as a list of strings (e.g., ["a","bc"]): '))

sol = Solution()
print("Are they equal?", sol.arrayStringsAreEqual(word1, word2))
```

## Rubber Duck Explanation

> “Each list is just pieces of a bigger word.
> I glue the pieces of `word1` together to get one string, and do the same for `word2`.
> If the two glued strings are exactly the same, I return `true`; otherwise, `false`.”

---
## Complexity

- **Time Complexity:** `O(N + M)` where `N` is total chars in `word1` and `M` in `word2` (string join + comparison).
- **Space Complexity:** `O(N + M)` to store the two concatenated strings.
